```python
from sklearn.datasets import load_iris
from sklearn.cross_decomposition import CCA
import matplotlib.pyplot as plt

# Wczytanie zbioru danych Iris
iris = load_iris()
X = iris.data  # Zestaw X zawiera pomiary cech irysów
Y = iris.target.reshape(-1, 1)  # Zestaw Y zawiera etykiety gatunków irysów

# Przeprowadzenie CCA
cca = CCA(n_components=1)
cca.fit(X, Y)

# Wypisanie macierzy kanonicznych współczynników
print("Macierz kanonicznych współczynników dla zestawu X:")
print(cca.x_loadings_)
print("\nMacierz kanonicznych współczynników dla zestawu Y:")
print(cca.y_loadings_)

# Obliczenie wartości korelacji kanonicznej
corrs = cca.transform(X, Y)
#print("\nWartość korelacji kanonicznej: ", corrs)

# Wizualizacja projekcji na pierwszą parę kanoniczną
proj_X, proj_Y = cca.transform(X, Y)
plt.scatter(proj_X, proj_Y)
plt.xlabel("Projekcja X")
plt.ylabel("Projekcja Y")
plt.title("Projekcje na pierwszą parę kanoniczną")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
link2data = "https://raw.githubusercontent.com/mwaskom/seaborn-data/master/penguins.csv"
df = pd.read_csv(link2data)
df =df.dropna()
df.head()
```

```python
X = df[['bill_length_mm','bill_depth_mm']]
X.head()
```

```python
X_mc = (X-X.mean())/(X.std())
X_mc.head()
```

```python
Y = df[['flipper_length_mm','body_mass_g']]
Y.head()
```

```python
Y_mc = (Y-Y.mean())/(Y.std())
Y_mc.head()
```

```python
	
from sklearn.cross_decomposition import CCA
```

```python
ca = CCA()
ca.fit(X_mc, Y_mc)
X_c, Y_c = ca.transform(X_mc, Y_mc)
```

```python
cc_res = pd.DataFrame({"CCX_1":X_c[:, 0],
                       "CCY_1":Y_c[:, 0],
                       "CCX_2":X_c[:, 1],
                       "CCY_2":Y_c[:, 1],
                       "Species":df.species.tolist(),
                      "Island":df.island.tolist(),
                      "sex":df.sex.tolist()})
```

```python
cc_res.head()
```

```python
import numpy as np
np.corrcoef(X_c[:, 0], Y_c[:, 0])
```

```python
np.corrcoef(X_c[:, 1], Y_c[:, 1])
```

```python
sns.set_context("talk", font_scale=1.2)
plt.figure(figsize=(10,8))
sns.scatterplot(x="CCX_1",
                y="CCY_1", 
                data=cc_res)
plt.title('Comp. 1, corr = %.2f' %
         np.corrcoef(X_c[:, 0], Y_c[:, 0])[0, 1])
```

```python
sns.set_context("talk", font_scale=1.2)
plt.figure(figsize=(10,8))
sns.scatterplot(x="CCX_2",
                y="CCY_2", 
                data=cc_res)
plt.title('Comp. 1, corr = %.2f' %
         np.corrcoef(X_c[:, 1], Y_c[:, 1])[0, 1])
```

```python
plt.figure(figsize=(10,8))
sns.scatterplot(x="CCX_1",
                y="CCY_1", 
                hue="Species", data=cc_res)
plt.title('First Pair of Canonical Covariate, corr = %.2f' %
         np.corrcoef(X_c[:, 0], Y_c[:, 0])[0, 1])
```

```python
plt.figure(figsize=(10,8))
sns.scatterplot(x="CCX_2",
                y="CCY_2", 
                hue="Species", data=cc_res)
plt.title('First Pair of Canonical Covariate, corr = %.2f' %
         np.corrcoef(X_c[:, 0], Y_c[:, 0])[0, 1])
```

```python
import numpy as np
from sklearn.decomposition import PCA
pca = PCA(n_components=2)

principal_components = pca.fit_transform(cc_res[['CCX_1', 'CCY_1', 'CCX_2', 'CCY_2']])

```

```python
principal_components.shape
```

```python
cc_res['pca_1'] = principal_components[:,0]
```

```python
cc_res['pca_2'] = principal_components[:,1]
```

```python
plt.figure(figsize=(10,8))
sns.scatterplot(x="pca_1",
                y="pca_2", 
                hue="Species", data=cc_res)
plt.title('First Pair of Canonical Covariate, corr = %.2f' %
         np.corrcoef(X_c[:, 0], Y_c[:, 0])[0, 1])
```
